## Bayesian Inference Calculation
To calculate the posterior probability that the patient has heart disease given a positive test result, we will use Bayes' Theorem:

$$
P(Disease | Positive) = \frac{P(Positive | Disease) \times P(Disease)}{P(Positive)}
$$

Given:
- $$P(Disease) = 0.02$$ (prior probability)
- $$P(Positive | Disease) = 0.95$$ (test sensitivity)
- $$P(Positive | No Disease) = 0.05$$ (false positive rate)

First, we need to calculate $$P(Positive)$$:
$$
P(Positive) = P(Positive | Disease) \times P(Disease) + P(Positive | No Disease) \times P(No Disease)
$$
Where $$P(No Disease) = 1 - P(Disease) = 0.98$$.

Calculating $$P(Positive)$$:
$$
P(Positive) = (0.95 \times 0.02) + (0.05 \times 0.98) = 0.019 + 0.049 = 0.068
$$

Now substituting back into Bayes' Theorem:
$$
P(Disease | Positive) = \frac{0.95 \times 0.02}{0.068} \approx \frac{0.019}{0.068} \approx 0.279
$$

**Answer:** Posterior Probability: **27.9%**

## Decision-Making Simulation
To calculate the expected health improvement for the patient based on treatment outcomes, we use the formula:

$$
\text{Expected Health Improvement} = P(Success) \times \text{Benefit} + P(Failure) \times (-\text{Loss})

= 70 - 15 = 55
