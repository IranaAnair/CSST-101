Part 1: Implement Basic Probability Calculations

```python
def get_probabilities():
    # Prompt the user to input probabilities
    p_A = float(input("Enter the probability of event A: "))
    p_B = float(input("Enter the probability of event B: "))
    p_B_given_A = float(input("Enter the probability of B given A: "))

    return p_A, p_B, p_B_given_A

def calculate_probabilities(p_A, p_B, p_B_given_A):
    #Joint Probability
    p_A_and_B = p_A * p_B

    #Marginal Probability
    p_a_given_b = p_A + p_B - p_A_and_B

    #Conditional Probability
    p_A_or_B = (p_B_given_A *p_A) / p_B

    return p_A_and_B, p_a_given_b, p_A_or_B

def display_results(p_A_and_B, p_a_given_b, p_A_or_B):
  print("\nResults:")
  print(f"1. Joint Probability: {p_A_and_B}")
  print(f"2. Marginal Probability: {p_a_given_b:.2f}")
  print(f"3. Conditional Probability: {p_A_or_B}")

def main():
    print("Probability Calculator")

    # Get input probabilities from user
    p_A, p_B, p_B_given_A = get_probabilities()

    # Calculate the required probabilities
    p_A_and_B, p_a_given_b, p_A_or_B = calculate_probabilities(p_A, p_B, p_B_given_A)

    # Display the results
    display_results(p_A_and_B, p_a_given_b, p_A_or_B)

if __name__ == "__main__":
    main()
```

Part 2: Bayesian Inference for Real-World Scenario

```python
def get_input():
    # Prompt the user to input probabilities
    prior = float(input("Enter the prior probability of disease: "))
    likelihood_positive_given_disease = float(input("Enter the likelihod of a positive test given disease: "))
    likelihood_positive = float(input("Enter the overall probability oof a positive test: "))

    return prior, likelihood_positive_given_disease, likelihood_positive

def bayesian_inference(prior, likelihood_positive_given_disease, likelihood_positive):
    # Calculate the posterior probability
    posterior = (likelihood_positive_given_disease * prior) / likelihood_positive
    return posterior

def display_inference(posterior):
    print(f"\nResults:")
    print(f"Posterior Probability: {posterior:.2f}")

def main():
    print("Bayesian Inference")
    prior,  likelihood_positive_given_disease, likelihood_positive = get_input()
    posterior = bayesian_inference(prior, likelihood_positive_given_disease, likelihood_positive)
    display_inference(posterior)

if __name__ == "__main__":
    main()
```

Part 3: Simulate Decision-Making Under Uncertainty

```python
import numpy as np

def get_inputs():
  p_succes = float(input("Enter the probability of success: "))
  reward = int(input("Enter the reward for success: "))
  penalty = int(input("Enter the penalty for failure: "))
  num_simulations = int(input("Enter the number of simulations: "))
  return p_succes, reward, penalty, num_simulations

def simulate_decision_making(p_succes, reward, penalty, num_simulations):
  outcomes = []
  for _ in range(num_simulations):
    if np.random.rand() < p_succes:
      outcomes.append(reward)
    else:
      outcomes.append(penalty)
  return np.mean(outcomes)

def display_simulations(outcomes):
  print("\nSimulation Results")
  print(f"Average Return: {outcomes}")

def main():
  print("Decision-Making Simulation")
  p_succes, reward, penalty, num_simulations = get_inputs()
  outcomes = simulate_decision_making(p_succes, reward, penalty, num_simulations)
  display_simulations(outcomes)

if __name__ == "__main__":
  main()
```

Part 4: Visualize Probability Distributions


```python
import matplotlib.pyplot as plt

# Function to simulate a binomial distribution
def simulate_binomial(n, p):
    # Simulate the number of heads in n_trials
    results = np.random.binomial(n=1, p=p, size=n)
    return results

# Function to simulate a normal distribution
def simulate_normal(n_trials, mean, std_dev):
    # Simulate exam scores based on a normal distribution
    results = np.random.normal(mean, std_dev, n_trials)
    return results


choice = input("Choose a distribution to simulate (binomial/normal): ").strip().lower()

if choice == "binomial":
    n = int(input("Enter the number of trials: "))
    p = float(input("Enter the probability of heads: "))
    data = simulate_binomial(n, p)
    plt.hist(data, bins=2)
    plt.title('Histogram of Binomial Distribution')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
elif choice == "normal":
    n = int(input("Enter the number of trials: "))
    mean = float(input("Enter the mean score: "))
    std_dev = float(input("Enter the standard deviation: "))
    data = simulate_normal(n, mean, std_dev)
    plt.hist(data, bins=30, density=True)
    plt.title('Histogram of Normal Distribution')
    plt.xlabel('Scores')
    plt.ylabel('Probability Density')
else:
    print("Invalid choice. Please choose 'binomial' or 'normal'.")

    plt.show()
```

Part 5: Real-World Scenario Prediction

```python
humidity = float(input("Enter the humidity level (0 to 1): "))
cloud_cover = float(input("Enter the cloud cover level (0 to 1): "))

def predict_rain(humidity, cloud_cover):
 # Define conditional probabilities based on your domain knowledge or data.
  # These are simplified examples.
  p_rain_given_high_humidity = 0.8
  p_rain_given_low_humidity = 0.2
  p_rain_given_high_cloud_cover = 0.7
  p_rain_given_low_cloud_cover = 0.3

  # Determine if humidity and cloud cover are high or low.
  high_humidity_threshold = 0.6
  high_cloud_cover_threshold = 0.6

  if humidity > high_humidity_threshold:
    p_rain_given_humidity = p_rain_given_high_humidity
  else:
    p_rain_given_humidity = p_rain_given_low_humidity

  if cloud_cover > high_cloud_cover_threshold:
    p_rain_given_cloud_cover = p_rain_given_high_cloud_cover
  else:
    p_rain_given_cloud_cover = p_rain_given_low_cloud_cover

  # Combine the probabilities using a simple approach (e.g., averaging).
  # More sophisticated methods can be used for combining evidence.
  predicted_probability_rain = (p_rain_given_humidity + p_rain_given_cloud_cover) / 2

  return predicted_probability_rain

probability_rain = predict_rain(humidity, cloud_cover)
print(f"The predicted probability of rain is: {probability_rain}")
```
